```python
x = np.linspace(-500, 500, 2000)
def sigmod(x):
    return 1.0 / (1.0 + np.exp(-x))
y = sigmod(x)
fig = plt.figure(figsize=(10,10))
ax1 = fig.add_axes([0.1,0.1,0.8,0.8])
ax1.plot(x,y,'orange')
ax1.set_title('sigmod(x)')
ax1.set_xlim(-5,5)
plt.annotate("(0,0.5)", xy = (0,0.5),  xytext = (2,0.4), arrowprops = dict(facecolor = 'blue', headlength = 10, headwidth = 10, width = 1), fontsize = 30)
plt.text(-4,0.5, "y=sigmod(x)", fontsize = 20, color = 'green')
ax2 = fig.add_axes([0.2,0.6,0.2,0.2])
ax2.plot(x,y,'red')
ax2.set_title('macro sigmod(x)')
ax2.set_xlim(-500,500)
plt.show()
```
