```python
import pandas as pd

train_set = pd.read_csv('data.txt', sep='	')

import numpy as np

label = np.array(train_set['fruit_label'])
train = pd.concat([train_set['mass'], train_set['width'], 
                   train_set['height'],train_set['color_score']],axis=1)
train = np.array(train)

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
train = scaler.fit_transform(train)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split

X_train, X_test,y_train,y_test = train_test_split(train,label,train_size = 0.6
                                                  ,random_state=1)
knn = KNeighborsClassifier(n_neighbors = 5, weights='distance')
knn.fit(X_train, y_train)
```

```python
from sklearn import*
y_pred = knn.predict(X_test)
print("model accuracy:", metrics.accuracy_score(y_test, y_pred))
```

```python
#加载数据
import pandas as pd
seed = pd.read_table('../seeds_dataset.txt',names=['A','P','C','LK','WK','AC','LKG','K'],sep='\s+')
#分离标签与特征
import numpy as np
slabel = np.array(seed.K)
strain = np.array(seed.drop(['K'],axis=1))
#标准化
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
strain = scaler.fit_transform(strain)
#加载KNN
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from math import sqrt
df = pd.DataFrame(columns = ['k','train_size','acc'])
i = 0
for ts in range(50,95,5):
    for k in range(1, int(sqrt(len(slabel)*ts/100.0)) + 9):
        acc = 0.0
        for r in range(1, 11):
            X_strain, X_stest,y_strain,y_stest = train_test_split(strain,slabel,train_size = ts/100,random_state = r)
            sknn = KNeighborsClassifier(n_neighbors = k, weights = 'distance')
            sknn.fit(X_strain, y_strain)
            y_spred = sknn.predict(X_stest)
            acc += metrics.accuracy_score(y_stest, y_spred)
        acc /= 10.0
        df.loc[i] = [k, ts/100, acc]
        i += 1
df = df.pivot('train_size','k','acc')
#作热力图
import seaborn as sns
import matplotlib.pyplot as plt
cmap = sns.cm.rocket_r
fig = plt.figure(figsize=(16,9))
fig.suptitle('distance')
sns.heatmap(df,annot=True,linewidths=2,fmt=".3f",cmap=cmap)
```

```python
df = pd.DataFrame(columns = ['train_size','Gaussian','Bernoulli','Drop1', 'Drop2', 'Drop3'])
#加载数据
import pandas as pd
seed = pd.read_table('../seeds_dataset.txt',names=['A','P','C','LK','WK','AC','LKG','K'],sep='\s+')
#分离标签与特征
import numpy as np
slabel = np.array(seed.K)
strain_1 = np.array(seed.drop(['K'],axis=1))
strain_2 = np.array(seed.drop(['K','P'],axis=1))
strain_3 = np.array(seed.drop(['K','P','LK','WK'],axis=1))
strain_4 = np.array(seed.drop(['K','LKG'],axis=1))
#加载高斯贝叶斯
from sklearn.naive_bayes import GaussianNB, BernoulliNB
from sklearn.model_selection import train_test_split
i = 0
for ts in range(50,95,5):
    Gacc = 0
    Bacc = 0
    GD1acc = 0
    GD2acc = 0
    GD3acc = 0
    for r in range(1, 11):
        
        X_train, X_test, y_train, y_test = train_test_split(strain_1, slabel, train_size=ts/100, random_state=r)
        Gmodel = GaussianNB()
        Gmodel.fit(X_train, y_train)
        y_pred = Gmodel.predict(X_test)
        Gacc += metrics.accuracy_score(y_test, y_pred)
        
        Bmodel = BernoulliNB()
        Bmodel.fit(X_train, y_train)
        y_pred = Bmodel.predict(X_test)
        Bacc += metrics.accuracy_score(y_test, y_pred)
        
        X_train, X_test, y_train, y_test = train_test_split(strain_2, slabel, train_size=ts/100, random_state=r)
        Gmodel = GaussianNB()
        Gmodel.fit(X_train, y_train)
        y_pred = Gmodel.predict(X_test)
        GD1acc += metrics.accuracy_score(y_test, y_pred)
        
        X_train, X_test, y_train, y_test = train_test_split(strain_3, slabel, train_size=ts/100, random_state=r)
        Gmodel = GaussianNB()
        Gmodel.fit(X_train, y_train)
        y_pred = Gmodel.predict(X_test)
        GD2acc += metrics.accuracy_score(y_test, y_pred)
        
        X_train, X_test, y_train, y_test = train_test_split(strain_4, slabel, train_size=ts/100, random_state=r)
        Gmodel = GaussianNB()
        Gmodel.fit(X_train, y_train)
        y_pred = Gmodel.predict(X_test)
        GD3acc += metrics.accuracy_score(y_test, y_pred)
    Gacc /= 10.0
    Bacc /= 10.0
    GD1acc /= 10.0
    GD2acc /= 10.0
    GD3acc /= 10.0
    df.loc[i] = [ts/100, Gacc, Bacc, GD1acc, GD2acc, GD3acc]
    i += 1

import matplotlib.pyplot as plt
plt.figure(figsize=(16,9))
plt.grid(alpha=0.7,linestyle=':')
plt.plot(df.train_size,df.Gaussian,color = 'purple',label = 'original')
plt.plot(df.train_size,df.Drop1,color = 'blue',label = 'Drop "P"')
plt.plot(df.train_size,df.Drop2,color = 'red',label = 'Drop MORE')
plt.plot(df.train_size,df.Drop3,color = 'green',label = 'Drop "LKG"')
plt.legend()
plt.xlabel('Train size')
plt.ylabel('Accuracy')
plt.show()
```

```python
#加载数据
import pandas as pd
seed = pd.read_table('../seeds_dataset.txt',names=['A','P','C','LK','WK','AC','LKG','K'],sep='\s+')
#分离标签与特征
import numpy as np
slabel = np.array(seed.K)
strain = np.array(seed.drop(['K'],axis=1))

from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
from sklearn import preprocessing
import pandas as pd

scale = pd.DataFrame(preprocessing.scale(strain))
# 3 * 4画布
fig = plt.figure(figsize=(16,12))
# 5 10 15 20
# 25 30 35 40
# 45 50 70 100
perplex = [5, 10, 15, 20, 25, 30, 35, 40, 45, 50, 70, 100]
# 绘制12个二维散点图
for i in range(1,13):
    model = TSNE(learning_rate=100, n_components=2, random_state=1, perplexity=perplex[i-1])
    ax = plt.subplot(3,4,i)
    ax.set_title("Perplexity " + str(perplex[i-1]))
    emmm = model.fit_transform(scale)
    plt.scatter(emmm[:, 0], emmm[:, 1], c=slabel)
```

```python
df = pd.DataFrame(columns = ['depth','train_size','acc'])
#加载数据
import pandas as pd
seed = pd.read_table('../seeds_dataset.txt',names=['A','P','C','LK','WK','AC','LKG','K'],sep='\s+')
#分离标签与特征
import numpy as np
slabel = np.array(seed.K)
strain = np.array(seed.drop(['K'],axis=1))
#标准化
from sklearn import preprocessing
strain = pd.DataFrame(preprocessing.scale(strain))
#加载决策树
i = 0
from sklearn import tree
for ts in range(50,95,5):
    for d in range(3, 11):
        acc = 0
        for r in range(1, 11):
            X_train, X_test, y_train, y_test = train_test_split(strain, slabel, train_size = ts/100, random_state = r)
            clf = tree.DecisionTreeClassifier(max_depth = d)
            clf.fit(X_train, y_train)
            y_pred = clf.predict(X_test)
            acc += metrics.accuracy_score(y_test, y_pred)
        acc /= 10.0
        df.loc[i] = [d, ts/100, acc]
        i += 1
df = df.pivot('train_size','depth','acc')
#作热力图
import seaborn as sns
import matplotlib.pyplot as plt
cmap = sns.cm.rocket_r
fig = plt.figure(figsize=(16,9))
fig.suptitle('Decision Tree Entropy')
sns.heatmap(df,annot=True,linewidths=2,fmt=".3f",cmap=cmap)
```
